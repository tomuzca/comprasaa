```python
import pandas as pd

# Leer los archivos me2l y mb51
me2l_df = pd.read_excel('ME2L_acmanet010224.xlsx')  
mb51_df = pd.read_excel('MB51acmanet0102.xlsx')  


merged_df = pd.merge(me2l_df, mb51_df, on='OC', how='inner')  


merged_df['DiasHabiles'] = (merged_df['Fecha de entrada'] - merged_df['Fecha de entrega']).dt.days


result_df = merged_df[['Nombre del proveedor', "OC",'Fecha de entrega','Fecha de entrada',"Texto breve","Cantidad de pedido", 'Ctd.en UM entrada', 'DiasHabiles']]

result_df.head()
```

```python
result_df.to_excel('resultado0102.xlsx', index=False)  
```

```python
def encontrar_filas_duplicadas(df):
    duplicadas = df[df.duplicated(keep=False)]
    return duplicadas


# Encontrar filas duplicadas
filas_duplicadas = encontrar_filas_duplicadas(result_df)

# Mostrar el resultado
print("Filas duplicadas en todas las columnas:")
print(filas_duplicadas)
```

```python
def eliminar_filas_duplicadas(df):
    # Encuentra las filas duplicadas en todas las columnas
    duplicadas = df[df.duplicated(keep=False)]

    # Elimina las filas duplicadas y crea un nuevo DataFrame
    df_sin_duplicados = df.drop_duplicates(keep=False)

    return duplicadas, df_sin_duplicados

filas_duplicadas, df_sin_duplicados = eliminar_filas_duplicadas(result_df)
```

```python
df_sin_duplicados.to_excel('eatotifacmanet.xlsx', index=False)  
```
