```python
import pandas as pd

# Leer los archivos me2l y mb51
me2l_df = pd.read_excel('ME2L250124.xlsx')  
mb51_df = pd.read_excel('mb512501.xlsx')  

# Fusionar los DataFrames en función de algún campo clave (por ejemplo, número de pedido)
merged_df = pd.merge(me2l_df, mb51_df, on='OC', how='inner')  # Ajusta el nombre del campo clave

# Calcular la diferencia de días hábiles
merged_df['DiasHabiles'] = (merged_df['Fecha de entrada'] - merged_df['Fecha de entrega']).dt.days

# Seleccionar las columnas necesarias
result_df = merged_df[['OC', 'Nombre del proveedor', 'Cantidad de pedido','Fecha de entrada', 'Fecha de entrega', "Texto breve", 'Ctd.en UM entrada', 'DiasHabiles']]

# Guardar el resultado en un nuevo archivo Excel
result_df.to_excel('resultado.xlsx', index=False)  # Ajusta el nombre del archivo según tus preferencias
```

```python
import pandas as pd

# Leer datos desde el archivo Excel
try:
    df = pd.read_excel("ME2L250124.xlsx")
except FileNotFoundError:
    print("El archivo ME2L250124.xlsx no se encontró.")
    exit()


# Convertir la columna "Fecha documento" a formato datetime
df["Fecha documento"] = pd.to_datetime(df["Fecha documento"], errors="coerce")


```

```python
import pandas as pd
from datetime import datetime



# Diccionario para almacenar la cantidad de OC por día
oc_por_dia = {}

# Procesar las órdenes de compra y contar la cantidad por día
for index, row in df.iterrows():
    fecha = row["Fecha documento"]
    cantidad = row["OC"]
    
    # Convertir la fecha a un objeto datetime para trabajar con ella
    try:
        fecha_obj = datetime.strptime(str(fecha), "%Y-%m-%d").date()
    except ValueError:
        print(f"Error en la fila {index + 2}: Fecha no válida - {fecha}")
        continue

    # Sumar la cantidad al día correspondiente en el diccionario
    if fecha_obj in oc_por_dia:
        oc_por_dia[fecha_obj] += cantidad
    else:
        oc_por_dia[fecha_obj] = cantidad

# Imprimir la cantidad de OC por día
for fecha, cantidad_oc in oc_por_dia.items():
    print(f"Fecha: {fecha}, Cantidad de OC: {cantidad_oc}")
```

```python
oc_por_fecha = df.groupby("Fecha documento")["OC"].nunique()

# Imprimir la cantidad de órdenes de compra diferentes por fecha
print(oc_por_fecha)
```

```python
# Crear una nueva columna con el número de la semana
df["Semana"] = df["Fecha documento"].dt.isocalendar().week


# Calcular la cantidad promedio de órdenes de compra únicas por semana
promedio_oc_unicas_por_semana = df.groupby("Semana")["OC"].nunique()

# Imprimir la cantidad promedio de órdenes de compra únicas por semana
print(promedio_oc_unicas_por_semana)
```
