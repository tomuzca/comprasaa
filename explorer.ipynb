```python
import pandas as pd
import matplotlib.pyplot as plt

xls = "../acma_acmanet/src/Acma_21_22_23.xlsx"

df = pd.read_excel(xls, sheet_name="DB")
```

```python
df.info()
```

```python
conteo_regiones = df['region'].value_counts()

porcentajes = (conteo_regiones / len(df)) * 100

plt.figure(figsize=(8, 8))
pie_chart = plt.pie(porcentajes, labels=None, autopct='%1.1f%%', startangle=90)

plt.legend(porcentajes.index, title='Regiones', bbox_to_anchor=(1, 0.5), loc="center left", fontsize=10)

plt.title('Porcentaje de Viajes a Regiones')
plt.show()
```

```python
conteo_ciudades = df['ciudad'].value_counts()

porcentajes = (conteo_ciudades / len(df)) * 100

plt.figure(figsize=(8, 8))
pie_chart = plt.pie(porcentajes, labels=None, autopct='%1.1f%%', startangle=90)

plt.legend(porcentajes.index, title='Ciudades', bbox_to_anchor=(1, 0.5), loc="center left", fontsize=10)

plt.title('Porcentaje de Viajes a Ciudades')
plt.show()
```

```python
sumas_por_mes = df.groupby('Mes')['total'].sum().reset_index()


print("Sumatoria de la columna 'total' por mes:")
print(sumas_por_mes)


plt.bar(sumas_por_mes['Mes'], sumas_por_mes['total'])
plt.title('Sumatoria por mes años 21,22 y 23')
plt.xlabel('Meses')
plt.ylabel('Suma Total')
suma_total_todos_meses = df['total'].sum()
print("\nSuma total de la columna 'total' de todos los meses:", suma_total_todos_meses)

plt.show()
```

```python
suma_total_todos_meses = df['total'].sum()
print("\nSuma total de la columna 'total' correspondiente a los años 21,22 y 23:", suma_total_todos_meses)
```

```python
df23 = df[df['Año'] == 2023]


sumas_por_mes = df23.groupby('Mes')['total'].sum().reset_index()


print("Sumatoria de la columna 'total' por mes:")
print(sumas_por_mes)


plt.bar(sumas_por_mes['Mes'], sumas_por_mes['total'])
plt.title('Sumatoria de la columna "total" por mes')
plt.xlabel('Meses')
plt.ylabel('Suma Total')
suma_total_todos_meses = df23['total'].sum()
print("\nSuma total de la columna 'total' de todos los meses:", suma_total_todos_meses)
plt.show()
```
