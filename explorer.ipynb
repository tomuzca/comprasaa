```python
import pandas as pd
import matplotlib.pyplot as plt

xls = "../acma_acmanet/Acma_21_22_23.xlsx"

df = pd.read_excel(xls, sheet_name="DB")
```

```python
df.info()
```

```python
conteo_regiones = df['region'].value_counts()

porcentajes = (conteo_regiones / len(df)) * 100

plt.figure(figsize=(8, 8))
pie_chart = plt.pie(porcentajes, labels=None, autopct='%1.1f%%', startangle=90)

plt.legend(porcentajes.index, title='Regiones', bbox_to_anchor=(1, 0.5), loc="center left", fontsize=10)

plt.title('Porcentaje de Viajes a Regiones')
plt.show()
```

```python
conteo_ciudades = df['ciudad'].value_counts()

porcentajes = (conteo_ciudades / len(df)) * 100

plt.figure(figsize=(8, 8))
pie_chart = plt.pie(porcentajes, labels=None, autopct='%1.1f%%', startangle=90)

plt.legend(porcentajes.index, title='Regiones', bbox_to_anchor=(1, 0.5), loc="center left", fontsize=10)

plt.title('Porcentaje de Viajes a Ciudades')
plt.show()
```

```python
from pandas.plotting import table
ax = plt.gca()
ax.axis('off')
tabla = table(ax, pd.DataFrame(porcentajes).T, loc='center', colWidths=[0.2]*len(porcentajes), cellLoc='center', rowLoc='center')
tabla.auto_set_font_size(False)
tabla.set_fontsize(10)
tabla.scale(1.2, 1.2)


total_ciudades = len(df)
print(f'Total de ciudades: {total_ciudades}')

      
tabla_porcentajes_ciudades = pd.crosstab(index=df['ciudad'], columns='Porcentaje').apply(lambda r: (r / total_ciudades) * 100, axis=1)
print("\nTabla de Porcentajes por Ciudad:")
print(tabla_porcentajes_ciudades)

        # Mostrar el gráfico
plt.title('Porcentaje de Regiones')
plt.show()
```
